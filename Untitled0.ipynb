```python
# step 3 : define target variable (Y) AND features (X)
x=data[['Experience Years']]

y=data['Salary']
```

```python
# reasons for double and single brackets in defining X AND Y is theire shape
x.shape    # X is a DataFrame and needs to retain its 2D structure
```

```python
y.shape    # y is a Series and can remain 1D (single bracket).
```

```python
# step 4: train test split
# import library
from sklearn.model_selection import train_test_split

```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,train_size=0.7,random_state=3)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
# step 5 : Select model for regression problem
# 1. linear regression model ---- ols method ( least squarred mwthod )
# 2. can be solved using various mAthematical approach --- probability, entropy, ensemble technique,
# geometry ( euclidiean distance)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
# step 6 : train the model ( fit model)
# this is the step where machine learning happens

model.fit(xtrain,ytrain)
```

```python
model.intercept_                 # intercept value of linear line y= c+ mx
```

```python
model.coef_                # slope of a line
```

```python
# step 7: prediction
y_pred=model.predict(xtest)
```

```python
# step 8 : Accuracy of the model
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(ytest,y_pred)
```
